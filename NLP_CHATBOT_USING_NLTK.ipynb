```python
    # Install and import
!pip install nltk -q
import nltk, random, re
from collections import Counter
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('wordnet', quiet=True)
```

```python
responses = {
    "hello": ["Hi there! ", "Hello! How can I help?", "Hey!"],
    "how are you": ["I'm doing great! You?", "All good here! "],
    "your name": ["I'm SimpleBot!", "Call me ChatBot "],
    "bye": ["Goodbye! ", "See you later!", "Bye! Take care"],
    "help": ["I can chat about: hello, name, weather, jokes, python, bye"],
    "weather": ["I can't check weather, but it's always sunny in code! "],
    "joke": ["Why do programmers prefer dark mode? Because light attracts bugs! "],
    "python": ["Python is awesome! Easy to learn, powerful to use! "],
    "time": ["I don't have a watch, but it's coding time! "],
    "thanks": ["You're welcome! ", "No problem!", "Anytime!"]
}

def get(text):
    text = text.lower().strip()
    words = nltk.word_tokenize(text)

    # Check for keywords in input
    for keyword in responses:
        if keyword in text or any(word in keyword for word in words):
            return random.choice(responses[keyword])

    return "I don't understand. Try: hello, help, joke, python, bye"
```

```python
print(" SimpleBot: Hi! Type 'bye' to exit")
while True:
    user = input("You: ").strip()
    if user.lower() in ['bye', 'exit', 'quit']:
        print(" SimpleBot: Goodbye! ")
        break
    print(" SimpleBot:", get(user))
```
